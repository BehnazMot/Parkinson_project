```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches

# === RReliefF Data ===
features_RRF_all = ['F1', 'F18', 'F5', 'F16', 'F3', 'F19', 'F17', 'F2', 'F13', 'F7', 'F15', 'F10', 'F12', 'F6', 'F9', 'F14', 'F4', 'F11', 'F8']
RRF_All = [0.0573, 0.0153, 0.0071, 0.0047, 0.0014, 0.0014, 0.0011, 0.0003, -0.0006, -0.0010, -0.0027, -0.0019, -0.0024, -0.0040, -0.0040, -0.0038, -0.0032, -0.0038, -0.0040]

features_RRF_pcc = ['F1', 'F10', 'F8', 'F7', 'F11', 'F6', 'F5', 'F3', 'F9', 'F4', 'F2']
RRF_PCC = [0.0297, 0.0047, 0.0003, -0.0003, -0.0009, -0.0014, -0.0015, -0.0019, -0.0030, -0.0020, -0.0040]

features_RRF_vif = ['F1', 'F7', 'F2', 'F5', 'F8', 'F4', 'F3', 'F6']
RRF_VIF = [0.0397, 0.0003, 0, -0.0005, -0.0051, -0.0020, -0.0062, -0.0096]

# === mRMR Data ===
features_mRMR_all = ['F1', 'F2', 'F16', 'F18', 'F17', 'F13', 'F19', 'F11', 'F15', 'F7', 'F10', 'F12', 'F6', 'F9', 'F14', 'F3', 'F4', 'F5', 'F8']
mRMR_All = [1.3366, 0.3698, 0.3090, 0.2387, 0.1324, 0.0868, 0.0755, 0.0511, 0.0510, 0.0442, 0.0416, 0.0392, 0.0344, 0.0342, 0.0227, 0, 0, 0, 0]

features_mRMR_pcc = ['F1', 'F8', 'F10', 'F9', 'F7', 'F11', 'F5', 'F3', 'F6', 'F4', 'F2']
mRMR_PCC = [1.3366, 0.2972, 0.2337, 0.1375, 0.0862, 0.0742, 0.0473, 0.0447, 0.0365, 0.0313, 0]

features_mRMR_vif = ['F1', 'F2', 'F5', 'F7', 'F4', 'F8', 'F3', 'F6']
mRMR_VIF = [1.2209, 0.2244, 0.2238, 0.1341, 0.0300, 0.0231, 0, 0]

def plot_feature_importance(features, scores, ax):
    bars = ax.bar(range(len(scores)), scores, color='skyblue', edgecolor='black')
    ax.set_xticks(range(len(features)))
    ax.set_xticklabels(features, rotation=45, ha='right', fontsize=9)
    ax.set_xlabel('Features')
    ax.set_ylabel('Importance Score')
    ax.grid(axis='y', linestyle='--', alpha=0.7)

    # Highlight top 3 features in orange
    top3_idx = np.argsort(scores)[-3:]
    for i in top3_idx:
        bars[i].set_color('orange')
        ax.text(i, scores[i] + max(scores)*0.05, features[i], ha='center', va='bottom',
                fontsize=8, color='darkred', fontweight='bold')

    # Add legend for each subplot
    highlight_patch = mpatches.Patch(color='orange', label='Top 3 Features')
    other_patch = mpatches.Patch(color='skyblue', label='Other Features')
    ax.legend(handles=[highlight_patch, other_patch], loc='upper right', fontsize=8)

# --- Plot RReliefF Figure ---
fig1, axs1 = plt.subplots(1, 3, figsize=(15, 5))
fig1.suptitle('', fontsize=16, fontweight='bold')

plot_feature_importance(features_RRF_all, RRF_All, axs1[0])
axs1[0].set_title('a) All Features')

plot_feature_importance(features_RRF_pcc, RRF_PCC, axs1[1])
axs1[1].set_title('b) PCC-Selected Features')

plot_feature_importance(features_RRF_vif, RRF_VIF, axs1[2])
axs1[2].set_title('c) VIF-Selected Features')

plt.tight_layout(rect=[0, 0, 1, 0.93])

# --- Plot mRMR Figure ---
fig2, axs2 = plt.subplots(1, 3, figsize=(15, 5))
fig2.suptitle('', fontsize=16, fontweight='bold')

plot_feature_importance(features_mRMR_all, mRMR_All, axs2[0])
axs2[0].set_title('a) All Features ')

plot_feature_importance(features_mRMR_pcc, mRMR_PCC, axs2[1])
axs2[1].set_title('b) PCC-Selected Features')

plot_feature_importance(features_mRMR_vif, mRMR_VIF, axs2[2])
axs2[2].set_title('c) VIF-Selected Features')

plt.tight_layout(rect=[0, 0, 1, 0.93])

plt.show()
```
